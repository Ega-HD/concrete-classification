```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the paths for the dataset
dataset_path = "./dataset_sample/"
pos_path = os.path.join(dataset_path, "Positive")
neg_path = os.path.join(dataset_path, "Negative")

# Count the total number of images in each class
print(f"Total cracked concrete images (Positive): {len(os.listdir(pos_path))}")
print(f"Total intact concrete images (Negative): {len(os.listdir(neg_path))}")

# Retrieve the file path of the first image in the Positive folder
sample_cracked_image = os.path.join(pos_path, os.listdir(pos_path)[0])

# Read and load the image
img = mpimg.imread(sample_cracked_image)

# Display the image using Matplotlib
plt.imshow(img)
plt.title("Sample of Cracked Concrete")
plt.axis('off')
plt.show()
```
