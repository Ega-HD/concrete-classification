```python
import tensorflow as tf
import os

# Define basic parameters
# ResNet and MobileNet typically use 224x224 image sizes
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
DATA_DIR = "./data_sample/"

print("Setting up Training Dataset...")
train_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2, # 20% data used for validation, 80% for training
    subset="training",
    seed=123,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE
)

print("\nSetting up Validation Dataset...")
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE
)

# Check the class names (should be 'Negative' and 'Positive')
class_names = train_dataset.class_names
print(f"\nClass names detected: {class_names}")
```
